#### Research and write a short summary on what the "Zero Trust" security model means and how it applies to an AI Engineer.

Zero trust security model applies to an AI engineer such that the AI engineer must know how to isolate the deployment service, access the trained model, what can read and write logs, read the training data, and delete the original training data. The principle of least privilege should be implemented to ensure that it will be easier to isolate issues and know which access has access to a certain tool or process. Further minimizing potential losses or damages.

By implementing the model, this ensures that every access level is accounted for and there is not risk of possible effect on other processes or roles when a certain access has been breached or is not performing well.


#### "What are the key trade-offs in adjusting a neural network's weights versus configuring cloud security settings?

Key tradeoffs include the complexity as well as interpretability as it is a blackbox and we will only know once both of the mentioned items have written the logs or if something happens, the immediateness of feedback wherein we know that the weights are performing well or if the cloud settings do work as both require running or testing again to check if the updatres work. Failures are also accounted for here as if there is a failure to update or check the weights if it has been adjusted deliberately or if it was attacked and if the cloud security settings haven't been configured properly, it will trigger a snowball of issues which may be hard to address in a way that the only way to fix it is to go back to a known state without issues.

## Week 20: MLOps: Monitoring & Observability

### Cybersecurity Theory: SIEM & Log Analysis
In cybersecurity, we use SIEM (Security Information and Event Management) systems. Think of a SIEM as a giant vacuum cleaner that sucks up logs from every server, firewall, and database in a company and looks for patterns.
- Log Analysis: The process of reviewing records of events. A log entry usually contains: Timestamp, IP Address, Action Taken, and Status.
- Anomaly Detection: Finding the "needle in the haystack." If a user usually logs in from New York at 9:00 AM, but suddenly logs in from London at 3:00 AM and tries to download 50GB of data, that is an anomaly.

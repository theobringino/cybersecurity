## Week 20: MLOps: Monitoring & Observability

### Cybersecurity Theory: SIEM & Log Analysis
In cybersecurity, we use SIEM (Security Information and Event Management) systems. Think of a SIEM as a giant vacuum cleaner that sucks up logs from every server, firewall, and database in a company and looks for patterns.
- Log Analysis: The process of reviewing records of events. A log entry usually contains: Timestamp, IP Address, Action Taken, and Status.
- Anomaly Detection: Finding the "needle in the haystack." If a user usually logs in from New York at 9:00 AM, but suddenly logs in from London at 3:00 AM and tries to download 50GB of data, that is an anomaly.

Model drift detection mirrors the process of detecting a security anomaly in such a way that you compare the new data to training data while in security, you comapre current traffic to normal user behavior. P-value is used to check if the drift happens while failure count is used for security breaches. Both metrics allows us to know and see if there is something really anomalous with either the model or the network. 

Both ways use the concept of observability to look for issues and using a dashboard or a tool to check this allows anyone who checks the data the convenience of just looking at spikes and real time data in a dashboard to ensure that there is no fatigue when checking.

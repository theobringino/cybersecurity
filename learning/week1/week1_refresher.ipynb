OSI Model

Application - where humans process data and information
Presentation - ensures data is in usable format
Session - ensures session is connected on both ends
Transport - transports the data from point a to point b
Network Layer - allows to transfer the data on a given network
Data Link - allows for data transfer between two connected nodes
Physical - hardware where data is stored or where it travels
